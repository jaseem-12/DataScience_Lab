** 4. Create a random list contain 10 elements with values between 40 and 80**

```python
import random
random_list=[]
n=10
for i in range(n):
  random_list.append(random.randint(40,80))
print(random_list)
```

** 5. Decrement the value of each element in the above list**

```python
for x in range(len(random_list)-1,-1,-1):
  random_list[x]=random_list[x]-1
  print(random_list[x])
```
